import requests
from skimage import io
import matplotlib.pyplot as plt

API_KEY = 'fYNGiCQv3hvjhzH3ol0CNydKP6NIwloShtDLcyYj'

rover_name = 'curiosity'

sol = 4102  # 4200
page = 1  # P√°gina inicial

def fetch_photos(rover_name, sol, page, api_key):
    url = f'https://api.nasa.gov/mars-photos/api/v1/rovers/{rover_name}/photos'
    params = {
        'sol': sol,
        'page': page,
        'api_key': api_key
    }
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Erro na requisi√ß√£o: {response.status_code}")
        return None

# C√¢meras espec√≠ficas
def plot_images_from_camera(photos, camera_name):
    for photo in photos:
        if photo['camera']['name'] == camera_name:
            img_url = photo['img_src']
            img = io.imread(img_url)
            
            title = f"P√°gina {page} | Camera: {photo['camera']['name']} | ID: {photo['id']}"
            
            plt.imshow(img)
            plt.title(title)
            plt.axis('off')  # Desativar os eixos
            plt.show()

# Busca de Imagens
while True:
    data = fetch_photos(rover_name, sol, page, API_KEY)
    
    if data and 'photos' in data:
        photos = data['photos']
        
        if not photos:
            print("Nenhuma foto encontrada nesta p√°gina. Fim da pesquisa.")
            break
        
        plot_images_from_camera(photos, 'NAVCAM')  # Exemplo de c√¢mera, voc√™ pode alterar
        plot_images_from_camera(photos, 'FHAZ')    # Outra c√¢mera
        plot_images_from_camera(photos, 'RHAZ')    # Outra c√¢mera
        
        page += 1
    else:
        print("Erro ou dados n√£o encontrados.")
        break
